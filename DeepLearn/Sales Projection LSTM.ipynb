```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train)
```

```python
scaled_train = scaler.transform(train)
```

```python
scaled_test =scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# remember length of batch must be less than test set if you want to create validation data generator
len(test)
```

```python
length =12 #makes sense to use 12 ..les than 18 and it is a full year
generator = TimeseriesGenerator(scaled_train,scaled_train,length=length,batch_size=1)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,LSTM
```

```python
n_features = 1
```

```python
model = Sequential()
model.add(LSTM(100,activation='relu',input_shape=(length,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

```python
validation_generator = TimeseriesGenerator(scaled_test,scaled_test,length=length,batch_size=1)
```

```python
model.fit_generator(generator,epochs=20,
                    validation_data=validation_generator,callbacks=[early_stop])
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
test_prediction =[]
first_eval_batch = scaled_train[-length:]
# reshape before feeding into network for prediction
current_batch = first_eval_batch.reshape(1,length,n_features)

for i in range (len(test)):
    # get predicition 1 timestamp ahead..[0] is subscripting for just prediction value in returned predict object
    current_pred = model.predict(current_batch)[0]
    # append current prediction
    test_prediction.append(current_pred)
    # update batch to include current prediction and drop the first value thus 
    #moving current batch one step into the future
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_prediction=scaler.inverse_transform(test_prediction)
```

```python
test['predictions']=true_prediction
```

```python
test.plot(figsize=(12,8))
```

### now forecasting into the future by training on entire dataset

```python
full_scaler = MinMaxScaler()
scaled_full_data = full_scaler.fit_transform(df)
```

```python
length = 12 
generator = TimeseriesGenerator(scaled_full_data,scaled_full_data,
                               length=length,batch_size=1)
```

```python
model = Sequential()
model.add(LSTM(100,activation='relu',input_shape=(length,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')

model.fit_generator(generator,epochs=7)
```

```python
forecast =[]

forecast_length= 12 #ideally your forecast length should not be longer than your test length

first_eval_batch = scaled_train[-length:]
# reshape before feeding into network for prediction
current_batch = first_eval_batch.reshape(1,length,n_features)

for i in range (forecast_length):
    
    # get predicition 1 timestamp ahead..[0] is subscripting for just prediction value in returned predict object
    current_pred = model.predict(current_batch)[0]
    
    # append current prediction
    forecast.append(current_pred)
    
    # update batch to include current prediction and drop the first value thus 
    #moving current batch one step into the future
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
forecast = full_scaler.inverse_transform(forecast)
```
